##Analyze the mobility of Pentacene OFET SVA under different solvent
The charge injuection in OFET is one of the most important issue. It can be divided into two parts: the charge go over the barrier at electrode/emiconductor barrier and bulk resistrance which is the organic semicondutor (OSC) region between electrode/emiconductor interface and conducting channel. To analysis the effect from the two part, first of all, the contact resistance is extracted with modified transmission line method (M-TLM).

#### Part I: Date import and clean

```python
import numpy as np
from numpy import *
#Read all the filemnames in the data folder.
import glob
import os

from pandas import *
import pandas as pd
#Creat an empty dataframe for data import form the txt files
Datatable = pd.DataFrame()

#Read all the file need in the folder one file for one loop
for filename in glob.iglob(os.path.join('/Users/sunhuabin/Desktop/Data/NoPMMA/','*C3*single.txt')):
    #print str(filename)
    #with open(filename) as f:
        #Read one file use the read_table for txt and read_csv for csv
        #then, change it into a dataframe with special colume names (content device ID and channel length)
        #Afterall, Append it to the 'Datatable' dataframe for future analysis
        # Step0: Get the Device ID of the files
        fname = str(filename)
        Namelistpath = fname.split('/')
        txtnames = Namelistpath[6]
        Namelistfile = txtnames.split(' ')
        txtname = Namelistfile[3]
        
        #Step1: Read the colums of data into a empty dataframe "data"
        data = pd.DataFrame()
        AllData = read_table(filename,header = 0, sep = '\t',)
        data = DataFrame(AllData, columns = ['Vbg','Id','gm','Ibg'])
        
        #Step2: Give DeviceID to each dataframes
        DeviceID = [txtname] * 162
        data['ID'] = DeviceID
        
        #Step3: Append datas into a file "Datatable"
        Datatable = Datatable.append(data)
        #Datatable = Datatable.append(data)
#print Datatable
```

```python
#Select the data need to use for ploting
Data150 = Datatable[Datatable['ID'] == '150u']
#We will plot the devices with channel lenght of  50u to 300u to plot curve as an example
import pylab
import matplotlib.pyplot as plt

IDList = ['50u', '100u', '150u', '200u' ,'250u' ,'300u']
# We can also use another method to generate list ex. xlabels = ['Ticklabel %i' % i for i in range(n)]

ColorList = ['k', 'r', 'c', 'm', 'y', 'g']
#http://matplotlib.org/api/colors_api.html

MarkerList = ['.', '.', '.', '.', '.', '.'] 
#http://matplotlib.org/api/markers_api.html

#set picture size
fig = plt.figure()
fig.set_size_inches(6,5)

for (DataID, Color, Marker) in zip(IDList, ColorList, MarkerList):
    Data = Datatable[Datatable['ID'] == DataID]
    
    x = Data[Data.columns[0]]
    y = Data[Data.columns[1]]
    
    plt.plot(x, abs(y), marker = Marker, markersize = 10, color = Color, linestyle = '-')
    
plt.title('Transfer of Different Channel Length', fontsize=20)
plt.xlabel('Gate Voltage', fontsize=20)
plt.ylabel('Drain Current (A)', fontsize=20)
plt.yscale('log')
plt.xticks(fontsize=15)
plt.yticks([10**-10, 10**-8, 10**-6], fontsize=15) #特别注意 python中指数操作为 **

plt.legend(IDList, loc='upper right',frameon=False)
# loc for location and frameon for the box around

#Below is the way to modified the label ticks and limits
#plt.ylim(( -3*10**-6,-10**-9))
#plt.yticks([10**(-9), 10**(-7), 10**(-5)],[r"$10^{-9}$",r"$10^{-7}$",r"$10^{-5}$"], fontsize=20)
#http://matplotlib.sourceforge.net/users/mathtext.html
#And we can use latex type format between $ and $  for example: ax.set_yticklabels([r"$\Delta$", r"$\Gamma$",r"$\Lambda$",r"$\Omega$",r"$\Phi$",r"$\Pi$",r"$\Psi$"])

plt.show()

fig.savefig('Test-transferall.png', dpi=300, bbox_inches="tight")
```
