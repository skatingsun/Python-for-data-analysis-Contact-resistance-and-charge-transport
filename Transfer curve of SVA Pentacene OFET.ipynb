Our first goal is to analysis the contact resistance, we use a method called modified transmission line method (M-TLM) to extract the contact resistance from the transfer curves of devices with different channel lengths. 

>A short description of M-TLM method is gven here:
>The M-TLM method is improved from traditional transmission line method (TLM). TLM is desgined for getting realiable contact resistance in transistors. The concept of TLM can be descripted as: assume the conductance of the channel is uniform, with the total resistance ~ channel length line extend to channel length = 0, the resistance here is contact resistacne. This explanation can be discribe as: 

>$$R_{total}\cdot W = R_{c}\cdot W + R_{channel}\cdot L$$

>where $R_{total}$ is the total resistance of the device, $R_{c}$ is the contact resistance, $L$ and $W$ are channel length and width, and $R_{channel}$ is the channel resistance of the device. Need to point out, the current space distribution is not uniform all over the channel and the physical length of the channel is not always precious, thus the contact resistance have more space-meaning than physical meaning in this method. 
>Yong et. al (10.1063/1.3479476) developed the TLM method. With a simple transform of the equation of TLM method:

>$$R_{total}\cdot W/L = R_{c}\cdot W/L + R_{channel}$$

>more precious contact resistance in low field (gate bias) region can be extrated. In simple words, the contact resistance extract from the slope value of fitting line between $R_{total}\cdot W/L$ and $1/L$ is more precious.

Thus, in order to get the contact resistance from the transfer data. We nee to do following steps:
* First, to extract threshold voltage (Vth) in each transfer curve, which is used to correct and normolize the gate bias on the transistor. 
* Second, use the corrected data to caculate the total resistance for each device at selected corrected gate voltages.
* Third, use the $R_{total}\cdot W/L$ ~ $1/L$ fitting to get the contact resistance at different corrected gate voltages.
